```python
import pandas as pd
import numpy as np
```

```python
df_data = pd.read_excel('Acadia - BI Intern Dashboarding Data Set Assessment.xlsx', sheet_name='Dept Data')
df_segments = pd.read_excel('Acadia - BI Intern Dashboarding Data Set Assessment.xlsx', sheet_name='Segments')
df_profiles = pd.read_excel('Acadia - BI Intern Dashboarding Data Set Assessment.xlsx', sheet_name='Profiles')
df_dept = pd.read_excel('Acadia - BI Intern Dashboarding Data Set Assessment.xlsx', sheet_name='Departments')
display(df_data.head())
display(df_segments.head())
display(df_profiles.head())
display(df_dept.head())
```

```python
display(df_data.isna().sum())
display(df_segments.isna().sum())
display(df_profiles.isna().sum())
display(df_dept.isna().sum())
```

```python
display(df_data.shape)
display(df_segments.shape)
display(df_profiles.shape)
display(df_dept.shape)
```

```python
df_final = df_data.merge(df_segments, how='left', on='SegmentID')
df_final = df_final.merge(df_dept, how='left', on='DepartmentID')
```

```python
df_final.head()
```

```python
df_final = df_final.merge(df_profiles, how='left', on='ProfileID')
df_final.head()
```

```python
df_final['Profile Description'] = df_final['Profile Description'].fillna("Unknown")
```

```python
df_final.head()
```

## Create new feature

### 1. Average Sales per Customer

```python
df_final['Avg_Sales_Per_Customer'] = df_final['Sales'] / df_final['Customers']
df_final.head()
```

### 2. Sales Growth (Year-over-Year %)

```python
# Ensure 'Year' is sorted and numeric
df_final = df_final.sort_values(['Department Description', 'Segment Description', 'Profile Description', 'Year'])

# Calculate Sales Growth
df_final['Sales_Growth'] = (
    df_final.groupby(['Department Description', 'Segment Description', 'Profile Description'])['Sales']
            .pct_change()
)
```

```python
df_final.head()
```

```python
df_final.head(10).describe()
```

```python
temp = df_final.groupby(['Department Description', 'Segment Description', 'Profile Description'])['Sales']
temp.head()
```

### 3. Customer Growth (Year-over-Year %)

```python
df_final['Customer_Growth'] = (
    df_final.groupby(['Department Description', 'Segment Description', 'Profile Description'])['Customers']
            .pct_change()
)
```

```python
df_final.head()
```

```python
df_final.head(10).describe()
```

### 4. High-Value Segment Flag


```python
threshold = df_final['Avg_Sales_Per_Customer'].quantile(0.75)
df_final['High_Value_Segment'] = df_final['Avg_Sales_Per_Customer'].ge(threshold).map({True:'Yes', False:'No'})
```

```python
df_final.head()
```

```python
df_final.describe()
```

```python
# Only multiply Sales_Growth and Customer_Growth by 100 for percentage representation
df_final['Sales_Growth'] = df_final['Sales_Growth'] * 100
df_final['Customer_Growth'] = df_final['Customer_Growth'] * 100

# Round to 2 decimal places
df_final['Avg_Sales_Per_Customer'] = df_final['Avg_Sales_Per_Customer'].round(2)
df_final['Sales_Growth'] = df_final['Sales_Growth'].round(2)
df_final['Customer_Growth'] = df_final['Customer_Growth'].round(2)
```

```python
df_final['Sales_Growth'] = df_final['Sales_Growth'].fillna(0)
df_final['Customer_Growth'] = df_final['Customer_Growth'].fillna(0)
```

```python
df_final.head()
```

```python
# Calculate total sales and customers by year
total_sales_by_year = df_final.groupby('Year')['Sales'].sum()
total_customers_by_year = df_final.groupby('Year')['Customers'].sum()

# Calculate overall growth %
overall_sales_growth = ((total_sales_by_year.loc[2023] - total_sales_by_year.loc[2022]) / total_sales_by_year.loc[2022]) * 100
overall_customer_growth = ((total_customers_by_year.loc[2023] - total_customers_by_year.loc[2022]) / total_customers_by_year.loc[2022]) * 100
```

```python
display(overall_sales_growth), display(overall_customer_growth)
```
