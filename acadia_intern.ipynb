```python
import pandas as pd
import numpy as np
```

```python
df_data = pd.read_excel('Acadia - BI Intern Dashboarding Data Set Assessment.xlsx', sheet_name='Dept Data')
df_segments = pd.read_excel('Acadia - BI Intern Dashboarding Data Set Assessment.xlsx', sheet_name='Segments')
df_profiles = pd.read_excel('Acadia - BI Intern Dashboarding Data Set Assessment.xlsx', sheet_name='Profiles')
df_dept = pd.read_excel('Acadia - BI Intern Dashboarding Data Set Assessment.xlsx', sheet_name='Departments')
display(df_data.head())
display(df_segments.head())
display(df_profiles.head())
display(df_dept.head())
```

```python
display(df_data.isna().sum())
display(df_segments.isna().sum())
display(df_profiles.isna().sum())
display(df_dept.isna().sum())
```

```python
display(df_data.shape)
display(df_segments.shape)
display(df_profiles.shape)
display(df_dept.shape)
```

```python
df_final = df_data.merge(df_segments, how='left', on='SegmentID')
df_final = df_final.merge(df_dept, how='left', on='DepartmentID')
```

```python
df_final = df_final.merge(df_profiles, how='left', on='ProfileID')
df_final.head()
```

```python
df_final['Profile Description'] = df_final['Profile Description'].fillna("Unknown")
```
